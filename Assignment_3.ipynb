```python
#Q6
# Q6
import pandas as pd

from google.colab import drive

path = '/content/Employees.csv'
df = pd.read_csv(path)


# a)
shape = df.shape
print("Shape of the DataFrame:", shape)

# b)

print("\nSummary of DataFrame:\n")
summary = df.info()

# c)
descriptive_stats = df.describe()
print("\nDescriptive Statistics:\n", descriptive_stats)

# d)
print("\nFirst 5 rows of the DataFrame:\n", df.head())
print("\nLast 3 rows of the DataFrame:\n", df.tail(3))

# e)
average_salary = df["Salary"].mean()
total_bonus = df["Bonus"].sum()
youngest_age = df["Age"].min()
highest_rating = df["Rating"].max()

print("\nAverage Salary:", average_salary)
print("Total Bonus:", total_bonus)
print("Youngest Employee's Age:", youngest_age)
print("Highest Performance Rating:", highest_rating)

# f)
df_sorted = df.sort_values(by="Salary", ascending=False)
print("\nDataFrame sorted by Salary:\n", df_sorted)

# g)
def performance_category(rating):
    if rating >= 4.5:
        return "Excellent"
    elif rating >= 4.0:
        return "Good"
    else:
        return "Average"

df["Performance"] = df["Rating"].apply(performance_category)
print("\nDataFrame with Performance Category:\n", df)

# h)
missing_values = df.isnull().sum()
print("\nMissing Values in DataFrame:\n", missing_values)

# i)
df.rename(columns={"Employee_ID": "ID"}, inplace=True)
print("\nDataFrame after renaming Employee_ID to ID:\n", df)

# j)
filtered_employees = df[(df["Years_of_Experience"] > 5) & (df["Department"] == "IT")]
print("\nFiltered Employees:\n", filtered_employees)

# k)
df["Tax"] = df["Salary"] * 0.10
print("\nDataFrame with Tax Column:\n", df)

# l)
df.to_csv("employees_modified.csv", index=False)
print("\nModified DataFrame saved to 'employees_modified.csv'")
```
