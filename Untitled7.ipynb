```python
#Assignment 7
#Q1
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
#i
roll_no=102317108
np.random.seed(roll_no)
sales=np.random.randint(1000,5000,size=(12,4))
months = ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec']
categories = ['Electronics', 'Clothing', 'Home & Kitchen', 'Sports']

df=pd.DataFrame(sales,columns=categories,index=months)
print(df)

#ii
print(df.head())
#Summary of output
print(df.describe())
#Total sales per category
print(df.sum())
#Total sales per month
print(df.sum(axis=1))
#Average sales growth per category
print(df.diff().mean())
# Adding column total sales and growth rate to above dataset
df['Total sales']=df.sum(axis=1)
#Per_change in growth rate from previous month
df['Growth rate']=df['Total sales'].pct_change()*100
print(df)

#Applying a randomized discount
if roll_no%2==0:
  df['Electronics']=df['Electronics']*0.1
else:
    df['Electronics']=df['Electronics']*0.15

#iii
plt.figure(figsize=(10,8))
for category in categories:
 plt.plot(df.index,df[category],label='Sales growth trend')
plt.title('Sales record')
plt.xlabel('Index')
plt.ylabel('Growth rate')
plt.grid()
plt.legend()
#iv
plt.figure(figsize=(9,8))
sns.boxplot(data=df[categories])
plt.xlabel('Index')
plt.ylabel('Growth rate')
plt.grid()
plt.legend()
```

```python
#Q4
def swap_list_elements(lst, pos1, pos2):
    temp = lst[pos1]
    lst[pos1] = lst[pos2]
    lst[pos2] = temp
    return lst

# Example usage:
my_list = [1, 2, 3, 4, 5]
print("Before swap:", my_list)
swap_list_elements(my_list, 1, 3)
print("After swap:", my_list)
```

```python
#Q5
def swap_set_elements(s, elem1, elem2):
    if elem1 in s and elem2 in s:
        lst = list(s)
        idx1, idx2 = lst.index(elem1), lst.index(elem2)
        lst[idx1], lst[idx2] = lst[idx2], lst[idx1]
        return set(lst)
    return s

my_set = {1, 2, 3, 4, 5}
print("Before swap:", my_set)
my_set = swap_set_elements(my_set, 2, 4)
print("After swap:", my_set)
```
